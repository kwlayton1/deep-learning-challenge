## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' .
application_df = application_df.drop(['EIN'], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at Name value counts for binning
application_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Define a cutoff value
cutoff_value = 6

# Calculate value counts for 'NAME'
counts = application_df['NAME'].value_counts()

# Identify application types to be replaced
application_types_to_replace = counts[counts < cutoff_value].index.tolist()

# Replace application types with counts < cutoff_value
application_df['NAME'] = application_df['NAME'].apply(lambda x: 'Other' if x in application_types_to_replace else x)

# Display value counts after replacement
value_counts_after_replacement = application_df['NAME'].value_counts()
print(value_counts_after_replacement)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Define a cutoff value
cutoff_value = 528

# Calculate value counts for 'APPLICATION_TYPE'
counts = application_df['APPLICATION_TYPE'].value_counts()

# Identify application types to be replaced
application_types_to_replace = counts[counts < cutoff_value].index.tolist()

# Replace application types with counts < cutoff_value
application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].apply(lambda x: 'Other' if x in application_types_to_replace else x)

# Display value counts after replacement
value_counts_after_replacement = application_df['APPLICATION_TYPE'].value_counts()
print(value_counts_after_replacement)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# Calculate value counts for 'CLASSIFICATION'
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Filter and display only counts greater than 1
filtered_counts = classification_counts[classification_counts > 1]
filtered_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Define a cutoff value
cutoff_value = 1883

# Calculate value counts for 'CLASSIFICATION'
counts = application_df['CLASSIFICATION'].value_counts()

# Identify classification types to be replaced
classifications_to_replace = counts[counts < cutoff_value].index.tolist()

# Replace classification types with counts < cutoff_value
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].apply(lambda x: 'Other' if x in classifications_to_replace else x)

# Display value counts after replacement
value_counts_after_replacement = application_df['CLASSIFICATION'].value_counts()
print(value_counts_after_replacement)
```

```python
# Split our preprocessed data into our features and target arrays
# Seperate the features, X,  from the target variable, y
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns='IS_SUCCESSFUL')
print(y.value_counts())
print(X.head())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
X = pd.get_dummies(X)
print(X.head())
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
print(X_train_scaled [0])
```

## Compile, Train and Evaluate the Model

```python
from tensorflow.keras.optimizers import Adam

# Define the model - deep neural net (number input features & hidden nodes for each layer)
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  10
hidden_nodes_layer2 =  10
hidden_nodes_layer3 =  10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="ReLU")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="ReLU"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="ReLU"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the model with the Adam optimizer
nn.compile(loss="binary_crossentropy", optimizer=Adam(learning_rate=0.001), metrics=["accuracy"])

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
